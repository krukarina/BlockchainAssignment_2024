Initial data cleaning

```python
raw_2019 = raw_2019.drop(raw_2019.index[:2])
# Assuming df is your DataFrame with the column names in the second row
# Use the second row as column names
raw_2019.columns = raw_2019.iloc[1]
# Drop the second row by its positional index (1)
raw_2019 = raw_2019.drop(raw_2019.index[1])
raw_2019 = raw_2019.drop(raw_2019.index[0])
# Reset the index
raw_2019 = raw_2019.reset_index(drop=True)


raw_2021 = raw_2021.drop(raw_2021.index[:2])
# Assuming df is your DataFrame with the column names in the second row
# Use the second row as column names
raw_2021.columns = raw_2021.iloc[1]
# Drop the second row by its positional index (1)
raw_2021 = raw_2021.drop(raw_2021.index[1])
raw_2021 = raw_2021.drop(raw_2021.index[0])
# Reset the index
raw_2021 = raw_2021.reset_index(drop=True)

raw_2023 = raw_2023.drop(raw_2023.index[:2])
# Assuming df is your DataFrame with the column names in the second row
# Use the second row as column names
raw_2023.columns = raw_2023.iloc[1]
# Drop the second row by its positional index (1)
raw_2023 = raw_2023.drop(raw_2023.index[1])
raw_2023 = raw_2023.drop(raw_2023.index[0])
# Reset the index
raw_2023 = raw_2023.reset_index(drop=True)
```
